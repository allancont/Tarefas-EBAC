1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Profissão- Cientista de Dados_M10_support material.csv', sep=',')
df['data_ref'] = pd.to_datetime(df['data_ref'], format='%Y-%m-%d')
````

````python
df.select_dtypes(include=['datetime', 'object']).head(0)
````

````python
plt.close('all')
figs, axs = plt.subplots(ncols=2,nrows=3,figsize=(15,15))

ax = sns.countplot(x='data_ref',  hue = 'sexo', data=df,ax=axs[0][0])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
#----------------------------------------------
ax = sns.countplot(x='data_ref',  hue = 'tipo_renda', data=df,ax=axs[0][1])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
#----------------------------------------------
ax = sns.countplot(x='data_ref',  hue = 'estado_civil', data=df,ax=axs[1][0])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
#----------------------------------------------
ax = sns.countplot(x='data_ref',  hue = 'tipo_residencia', data=df,ax=axs[1][1])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
#----------------------------------------------
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=df,ax=axs[2][0])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
#----------------------------------------------
df2 = df[df.educacao.isin(['Secundário','Superior completo'])]
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=df2,ax=axs[2][1])
tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
labels = ax.set_xticklabels(tick_labs, rotation=90)
````

### Conclusão
De todas as variáveis qualitativas testadas, 'educacao' foi a que apresentou uma maior variação ao longo do tempo a partir de setembro/2015, merecendo, portanto, um aprofundamento na investição das causas deste fenômeno.
