```python
sinasc_raw = pd.read_csv('SINASC_RO_2019.csv')
sinasc_raw.head()
```

```python
sinasc_1 = sinasc_raw[['IDADEMAE', 'IDADEPAI','munResNome']]
```

# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
#idademedia = sinasc_1.groupby(['munResNome']).aggregate([('total','count'),('média','mean'),('máximo', 'max'),('mínimo','min')])
idademedia = sinasc_1.groupby(['munResNome']).mean()
idademedia['DIFERENCA']=idademedia.apply(lambda x: x['IDADEPAI']-x['IDADEMAE'],axis=1)
idademedia.sort_values('DIFERENCA',ascending=False)
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_2 = sinasc_raw[['ESCMAE', 'DTNASC','PESO','SEXO']]

sinasc_2[sinasc_2['DTNASC'] == '2019-05-27'].groupby(['ESCMAE','SEXO'])['PESO'].mean().unstack()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

### Município que nasceu menos bebês em 2019

```python
sinasc_raw[sinasc_raw['munResNome'] != 'Município ignorado - RO'].groupby('munResNome')['IDADEMAE'].count().sort_values(ascending=True).head(1)
```

### Idade média, máxima, mínima das mães


```python
sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras'].groupby('munResNome')['IDADEMAE'].aggregate([('média','mean'),('máximo', 'max'),('mínimo','min')])
```

### Idade média, máxima, mínima dos pais

```python
sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras'].groupby('munResNome')['IDADEPAI'].aggregate([('média','mean'),('máximo', 'max'),('mínimo','min')])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



### Município que nasceu mais bebês no mês de março

```python
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_3=sinasc_raw[sinasc_raw['DTNASC'].isin(pd.date_range('2019-03-01', '2019-03-31'))]
sinasc_3.groupby('munResNome')['IDADEMAE'].count().sort_values(ascending=False).head(1)
```

### Quantidade de filhos vivos média, máxima, mínima nesse municipio

```python
sinasc_3[sinasc_3['munResNome'] == 'Porto Velho'].groupby('munResNome')['QTDFILVIVO'].aggregate([('total','count'),('média','mean'),('máximo', 'max'),('mínimo','min')])
```

### Idade média, máxima, mínima dos pais nesse municipio

```python
sinasc_3[sinasc_3['munResNome'] == 'Porto Velho'].groupby(['munResNome'])['IDADEMAE','IDADEPAI'].aggregate([('total','count'),('média','mean'),('máximo', 'max'),('mínimo','min')])
```
