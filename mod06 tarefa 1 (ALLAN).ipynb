#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sinasc = pd.read_csv('SINASC_RO_2019.csv')
%matplotlib inline
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df=sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
df['APGAR1'].dropna().value_counts().sort_index().plot.bar(rot=30,subplots=True,color='red')
```

```python
df['APGAR5'].dropna().value_counts().sort_index().plot.bar(rot=30,subplots=True,color='green')
```

```python
df['SEXO'].dropna().value_counts().sort_index().plot.bar(rot=30,subplots=True,color='teal')
```

```python
df['GESTACAO'].dropna().value_counts().sort_index().plot.bar(rot=30,subplots=True,color='purple')
```

```python
df['ESTCIVMAE'].dropna().value_counts().sort_index().plot.bar(rot=30,subplots=True,color='blue')
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)
sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc)

#ajustar o espaçamento
plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
