# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import math
from scipy.stats import norm, t
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
media = 15
desvio_padrao = 0.1
target = 15.2
distribuicao_normal = norm(media, desvio_padrao)
probabilidade = distribuicao_normal.sf(target)

print(f"A probabilidade de um pacote ter mais que 15,2 kg é de {probabilidade:.2%}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
media = 15
desvio_padrao = 0.1

distribuicao_normal = norm(media, desvio_padrao)

percentil_95 = distribuicao_normal.ppf(0.95)

amostra = distribuicao_normal.rvs(size=10000)

sns.kdeplot(amostra)
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.axvline(percentil_95, color='r', linestyle='--', label='95%')
plt.legend()
plt.show()

print(f"O percentil de ordem 95% para o peso dos pacotes é de {percentil_95:.2f} kg")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Cria objeto Y
Y = np.linspace(-6, 6, 200)

# Distribuição Normal Padrão
dist_norm = norm()

# Probabilidades para a distribuição Normal Padrão
prob_norm = dist_norm.cdf(Y)

# Distribuição t-student com 5 graus de liberdade
dist_t = t(df=5)

# Probabilidades para a distribuição t-student com 5 graus de liberdade
prob_t = dist_t.cdf(Y)

# Imprime as probabilidades para cada valor em Y
for i in range(len(Y)):
    print(f"Para Y={Y[i]:.2f}: P(Y1 < {Y[i]:.2f}) = {prob_norm[i]:.4f}, P(Y2 < {Y[i]:.2f}) = {prob_t[i]:.4f}")
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
plt.plot(Y, prob_norm, label='Normal')
plt.plot(Y, prob_t, label='t-student')
plt.title('Probabilidade acumulada para Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade acumulada')
plt.legend()
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Criando o objeto com os valores de P
P = np.linspace(0, 1, 20)

# Distribuição Normal Padrão
dist_norm = norm()

# Percentis para a distribuição Normal Padrão
percentis_norm = [dist_norm.ppf(p) for p in P]

# Distribuição t-student com 5 graus de liberdade
# dist_t = t(df=5)
dist_t = stats.t(df=5)


# Percentis para a distribuição t-student com 5 graus de liberdade
percentis_t = [dist_t.ppf(p) for p in P]

# Impressão dos resultados
print("Percentis para Y1 ~ N(0, 1):", percentis_norm)
print("Percentis para Y2 ~ t(5):", percentis_t)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Criação do gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Adiciona as curvas de distribuição
ax.plot(percentis_norm, P, label='Normal')
ax.plot(percentis_t, P, label='t(5)')

# Adiciona as legendas
ax.set_xlabel('Percentis')
ax.set_ylabel('Probabilidades')
ax.set_title('Distribuições de Y1 e Y2')
ax.legend()

# Mostra o gráfico
plt.show()
```
